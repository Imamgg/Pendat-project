```python
from itertools import islice
from youtube_comment_downloader import *
import pandas as pd
import datetime
from tqdm import tqdm  # Untuk progress bar

# Inisialisasi downloader
downloader = YoutubeCommentDownloader()

# URL video yang akan di-scrape
url = "https://www.youtube.com/watch?v=Cqxfjuj-zPw&t=526s"

# Jumlah komentar yang ingin diambil
jumlah_komentar = 8000

# Mendapatkan komentar dari URL
comments = downloader.get_comments_from_url(url, SORT_BY_RECENT) 

# List untuk menyimpan komentar
comments_list = []

print(f"\nðŸ”„ Mengambil {jumlah_komentar} komentar teratas dari video...\n")

# Progres real-time dengan tqdm
for comment in tqdm(
    islice(comments, jumlah_komentar),
    total=jumlah_komentar,
    desc="ðŸ“¥ Mengunduh Komentar",
):
    comments_list.append(comment)

# Membuat DataFrame
df = pd.DataFrame(comments_list)

# Timestamp dan nama file
timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
filename = f"youtube_comments_{timestamp}.csv"

# Simpan sebagai CSV
df.to_csv(filename, index=False, encoding="utf-8-sig")

print(f"\nâœ… Komentar berhasil disimpan ke file: {filename}")
```
